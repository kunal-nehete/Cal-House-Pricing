```python
X=df.drop(columns='Price')
y=df.Price
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
#internally we use gradient descent our aim is to come close global minima
# to make sure to come to global minima 
```

#### Assumptions of Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
print(lr.coef_)
```

```python
import statsmodels.api as sm 
```

```python
# Let's define a function for the multiple regression
def OLS_Regression_Stats(x,y):
    
    x = sm.add_constant(x)
    
    # defining the model, fitting the model and printing the model summary
    sm_model = sm.OLS(y,x).fit()
    print(sm_model.summary())
```

```python
OLS_Regression_Stats(X_train,y_train)
```

```python
print(lr.intercept_)
print(lr.get_params())
```

```python
test_pred=lr.predict(X_test)
```

```python
test_pred
```

```python
plt.scatter(y_test,test_pred)
```

```python
train_pred=lr.predict(X_train)
```

##### Evaluation of model

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
print('MSE',mean_squared_error(y_test,test_pred))
print('RMSE',np.sqrt(mean_squared_error(y_test,test_pred)))
print('MAE',mean_absolute_error(y_test,test_pred))
print('MAPE',mean_absolute_percentage_error(y_test,test_pred),'%')
```

```python
# R2
from sklearn.metrics import r2_score
r2=r2_score(y_test,test_pred)
print(r2)
```

```python
# Adj R2
adj_r2=1-(1-r2)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
adj_r2
```
